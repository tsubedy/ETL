```python
import numpy as np
import pandas as pd
import sqlalchemy

# Python SQL toolkit and Object Relational Mapper
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from config import username, password
```

```python
from sqlalchemy import create_engine
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/county_covid_database')
connection = engine.connect()
```

```python
# Loading population data (using sql view 'population')
population = pd.read_sql("SELECT * from population", connection)
population.head()
```

```python
# Loading tests data (sql view 'tests')
tests = pd.read_sql("SELECT * from tests", connection)
tests.head()
```

```python
# loading cases data
cases = pd.read_sql("SELECT * from cases", connection)
cases.head()
```

```python
# Loading vaccination data
vaccination = pd.read_sql("SELECT * from vac", connection)
vaccination.head()
```

```python
# Loading combined dataset (using sql view 'county_wise_combined_data')

combined_data = pd.read_sql("SELECT * from county_wise_combined_data", connection)
combined_data.head()
```

```python
# County Population with higher density (population >= 1000 per sq miles) 

dense_county_data = pd.read_sql("SELECT * from high_pop_density", connection)
dense_county_data
```

```python
# Denser county full dataset

dense_county_full_data = pd.read_sql("select a.* from county_wise_combined_data a join high_pop_density h on a.county = h.county", connection)
dense_county_full_data
```
